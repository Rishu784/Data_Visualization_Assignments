```python
import numpy as np 
import pandas as pd
from datetime import datetime
```

```python
cust = pd.read_csv(r"C:\Users\susha\Documents\AWCustomers.csv")
cust
```

```python
sales = pd.read_csv(r"C:\Users\susha\Documents\AWSales.csv")
sales
```

```python
aw = cust.merge(sales, on="CustomerID", how="inner")
aw
```

```python
aw = aw.drop(["Title","MiddleName","Suffix","AddressLine2"],axis=1)
aw
```

```python
aw["BirthDate"] = pd.to_datetime(aw["BirthDate"])
aw["LastUpdated"] = pd.to_datetime(aw["LastUpdated"])
aw
```

```python
ref_date = pd.to_datetime("2014-01-01")
aw["Age"] = ((aw["LastUpdated"].fillna(ref_date) - aw["BirthDate"]) / np.timedelta64(1, "Y"))
aw
```

```python
aw_clean = aw.dropna(subset=[
    "BikeBuyer","YearlyIncome","Age",
    "MaritalStatus","Gender","Education","Occupation",
    "HomeOwnerFlag","NumberCarsOwned","NumberChildrenAtHome","TotalChildren"
])
aw_clean
```

```python
df = aw_clean[[
    "BikeBuyer","Age","YearlyIncome","MaritalStatus","Gender","Education",
    "Occupation","HomeOwnerFlag","NumberCarsOwned",
    "NumberChildrenAtHome","TotalChildren","AvgMonthSpend"
]]
df.head()
```
