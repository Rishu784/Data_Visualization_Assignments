```python
# Question 1->

import numpy as np
from scipy.spatial import distance
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import jaccard_score

customer_A = [4,5,2,3,4]
customer_B = [5,3,2,4,5]

customer_A_binary = [1, 0, 1, 1, 0, 1] 
customer_B_binary = [1, 1, 1, 0, 0, 1] 

# Euclidean Distance
Euc_dis = distance.euclidean(customer_A,customer_B)

# Manhattan Distance
manhattan_distance = distance.cityblock(customer_A,customer_B)

# Hamming Distance 
Ham_dis = distance.hamming(customer_A_binary,customer_B_binary) * len(customer_A_binary)

# Cosine Similarity
cos_sim = cosine_similarity([customer_A],[customer_B])[0][0]

# jaccard score
jac_score = jaccard_score(customer_A_binary,customer_B_binary)

print("Customer Preference Comparision : \n")
print(f"Euclidean Distance : {Euc_dis:.2f}")
print(f"Manhattan Distance : {manhattan_distance:.2f}")
print(f"Hamming Distance : {Ham_dis:.2f}")
print(f"Cosine Similarity : {cos_sim:.2f}")
print(f"Jaccard Score : {jac_score:.2f}")
```

```python
# Q2->

import numpy as np
from scipy.spatial import distance

user1 = [5,3,4,4,2]
user2 = [4,2,5,4,3]

cheb_dis = distance.chebyshev(user1,user2)
minkow_dis = distance.minkowski(user1,user2,p=3)

print("User Rating Comparison : \n")
print(f"Chebyshev Distance : {cheb_dis:.2f}")
print(f"Minkowski Distance : {minkow_dis:.2f}")
```
